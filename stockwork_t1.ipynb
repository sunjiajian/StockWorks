```python
zixuan_codes = pd.DataFrame([['上指', 'sh'], ['深指', 'sz'], ['创指', 'cyb'], ['创业板B', '150153'], ['互联网B', '150195'], ['煤炭B', '150290'], ['钢铁B', '150288'], ['证券B', '150172'], ['房地产B', '150118'], ['国防B', '150206'], ['克明面业', '002661'], ['第一创业', '002797'], ['万科A', '000002'], ['恒生电子', '600570'], ['天顺股份', '002800'], ['深物业A', '000011'], ['龙洲股份', '002682'], ['浙江世宝', '002703']])
def get_realtime_info():
    rti = ts.get_realtime_quotes(zixuan_codes[1])
    rti['percent'] = [100.0 * (float(rti['price'][i]) - float(rti['pre_close'][i]))/float(rti['pre_close'][i]) for i in range(len(rti.index))]
    rti['amount'] = [float(rti['amount'][i]) / 100000000.0 for i in range(len(rti.index))]
    return rti[['name','code','price','percent','high','low','amount','time']]
```

```python
def code2str(cd):
    if cd < 10:
        scd = '00000%d' % cd
        return scd
    elif cd < 100:
        scd = '0000%d' % cd
        return scd
    elif cd < 1000:
        scd = '000%d' % cd
        return scd
    elif cd < 10000:
        scd = '00%d' % cd
        return scd
    elif cd < 100000:
        scd = '0%d' % cd
        return scd
    else:
        scd = str(cd)
        return scd
```

```python
# zf>=4 and highper>=1 and lowper<=-1
def getGoodDays(stockcode, startdate, enddate):
    data = ts.get_k_data(stockcode, start=startdate, end=enddate)
    data['zf'] = [(float(data['high'][i]) - float(data['low'][i])) / float(data['close'][i-1]) * 100 if i > 0 else np.nan for i in range(len(data.index))]
    data['high_per'] = [float(data['high'][i]) / float(data['close'][i-1]) * 100 - 100 if i > 0 else np.nan for i in range(len(data.index))]
    data['low_per'] = [float(data['low'][i]) / float(data['close'][i-1]) * 100 - 100 if i > 0 else np.nan for i in range(len(data.index))]
    data['good_t_day'] = [True if data['zf'][i] >= 4 and data['high_per'][i] >= 1 and data['low_per'][i] <= -1 else False for i in range(len(data.index))]
    data['buy_sell_good'] = False
    data['sell_buy_good'] = False
    for i in range(1, len(data.index)):
        if data['good_t_day'][i]:
            ticks = ts.get_tick_data(code=stockcode, date=data['date'][i])
            if len(ticks.index) <= 0:
                continue
            lasthigh = -1
            lastlow = -1
            firsthigh = -1
            firstlow = -1
            for j in range(len(ticks.index)):
                if (float(ticks['price'][j]) / float(data['close'][i-1]) * 100 - 100) >= 1 and lasthigh == -1:
                    lasthigh = j
                if (float(ticks['price'][j]) / float(data['close'][i-1]) * 100 - 100) <= -1 and lastlow == -1:
                    lastlow = j
                if lasthigh >= 0 and lastlow >=0:
                    break
            tlen = len(ticks.index) - 1
            while tlen >= 0:
                if (float(ticks['price'][j]) / float(data['close'][i-1]) * 100 - 100) >= 1 and firsthigh == -1:
                    firsthigh = tlen
                if (float(ticks['price'][j]) / float(data['close'][i-1]) * 100 - 100) <= -1 and firstlow == -1:
                    firstlow = tlen
                if firsthigh >= 0 and firstlow >=0:
                    break
                tlen -= 1
            data['sell_buy_good'][i] = firsthigh < lastlow
            data['buy_sell_good'][i] = firstlow < lasthigh
    return data
```

```python
sbdata = getGoodDays(stockcode='002703', startdate='2016-08-01', enddate='2016-11-15')
sbdata
```
